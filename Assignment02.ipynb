## Problem 1

```
# Create a list of 5 numbers and print the sum of all numbers.

numbers=[10,20,30,40,50]
total=0
for num in numbers:
    total+=num
print('Sum =', total)
```

## Problem 3

```
# Count how many numbers are greater than 10.

numbers=[5,12,18,7,25,3,11]
count=0
for num in numbers:
    if num>10:
        count+=1
print('Numbers greater than 10 =', count)
```

## Problem 14

```
# Nested if positive/negative even/odd.
number=int(input('Enter a number: '))
if number>0:
    print('Positive')
    if number%2==0:
        print('Even')
    else:
        print('Odd')
else:
    print('Negative')
```

## Problem 15

```
# Pass/fail and grade.
marks=int(input('Enter marks: '))
if marks>=40:
    print('Pass')
    if marks>=80:
        print('Grade A')
    elif marks>=60:
        print('Grade B')
    else:
        print('Grade C')
else:
    print('Fail')
```

## Problem 16

```
# Adult and senior citizen.
age=int(input('Enter age: '))
if age>18:
    print('Adult')
    if age>60:
        print('Senior Citizen')
    else:
        print('Not a Senior Citizen')
else:
    print('Not an Adult')
```

## Problem 17

```
# Print numbers from 1 to n.
def print_numbers(n):
    for i in range(1,n+1):
        print(i)

print_numbers(5)
print_numbers(10)
```

## Problem 18

```
# Count down from n to 1.
def count_down(n):
    for i in range(n,0,-1):
        print(i)

count_down(5)
count_down(10)
```

## Problem 19

```
# Sum of list using function.
def sum_of_list(numbers):
    total=0
    for num in numbers:
        total+=num
    return total

print(sum_of_list([1,2,3,4,5]))
print(sum_of_list([10,20,30]))
```

## Problem 20

```
# Multiplication table.
def print_table(number):
    for i in range(1,11):
        print(f'{number} x {i} = {number*i}')

print_table(5)
print()
print_table(7)
```
